**1.1 Dataset pour Paris**

```python
#https://files.data.gouv.fr/geo-dvf/latest/csv/2016/departements/
link = "https://files.data.gouv.fr/geo-dvf/latest/csv/2016/departements/75.csv.gz"
table = pd.read_csv(link, sep = ",")

for year in range(2017,2022):
    name = "https://files.data.gouv.fr/geo-dvf/latest/csv/" + str(year) + "/departements/75.csv.gz"
    table = pd.concat([table, pd.read_csv(name, sep = ',')])

display("Taille de table :")
display(table.shape)
table.head()
```
